```python
#One operation using SciPy
```

```python
d1 = np.array(df['Horsepower'])
d2 = np.array(df['Torque (lb-ft)'])
correlation_coefficient, p_value = stats.pearsonr(d1, d2)
print(f"Data X: {d1}")
print(f"Data Y: {d2}")
print(f"Pearson Correlation Coefficient (r): {correlation_coefficient:.4f}")
print(f"2-tailed P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print(f"Conclusion: The correlation is statistically significant at the {alpha*100}% level.")
else:
    print(f"Conclusion: The correlation is NOT statistically significant at the {alpha*100}% level.")
```

```python
#One operation using Statsmodels
```

```python
independent_variable = 'Engine Size (L)'
dependent_variable = 'Horsepower'
X = sm.add_constant(df[independent_variable])
y = df[dependent_variable]
model = sm.OLS(y, X)
results = model.fit()
print("--- Linear Regression Results ---")
print(results.summary())
```

```python
#Correlation plot with a regression line.
```

```python
x_column = df['Engine Size (L)']
y_column = df['Torque (lb-ft)']
sns.regplot(x=x_column, y=y_column, data=df,
                ci=95, # Show 95% confidence interval for the regression estimate
                line_kws={'color': 'red', 'linestyle': '--', 'linewidth': 2},
                scatter_kws={'alpha': 0.6, 's': 50, 'color': 'blue'}) 
plt.title(f'Correlation between {x_column} and {y_column}', fontsize=16)
plt.xlabel(x_column, fontsize=12)
plt.ylabel(y_column, fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
